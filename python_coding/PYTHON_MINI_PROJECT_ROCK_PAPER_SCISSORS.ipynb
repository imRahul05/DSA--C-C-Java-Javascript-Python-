```python
import random
import tkinter as tk
from tkinter import ttk

class RockPaperScissorsGame:
    def __init__(self, window):
        self.window = window
        self.user_score = 0
        self.computer_score = 0
        self.choices = ['rock', 'paper', 'scissors']

        self.user_choice = tk.StringVar()
        self.computer_choice = tk.StringVar()
        self.result_text = tk.StringVar()

        self.create_widgets()

    def create_widgets(self):
        button_frame = ttk.Frame(self.window)
        button_frame.pack(pady=20)

        for choice in self.choices:
            button = ttk.Button(button_frame, text=choice, command=lambda choice=choice: self.play(choice))
            button.pack(side=tk.LEFT, padx=10)

        result_label = ttk.Label(self.window, textvariable=self.result_text, font=('Arial', 14))
        result_label.pack(pady=20)

        score_frame = ttk.Frame(self.window)
        score_frame.pack()

        user_score_label = ttk.Label(score_frame, textvariable=self.user_choice, font=('Arial', 12))
        user_score_label.pack(side=tk.LEFT, padx=10)

        computer_score_label = ttk.Label(score_frame, textvariable=self.computer_choice, font=('Arial', 12))
        computer_score_label.pack(side=tk.RIGHT, padx=10)

        # Restart button
        self.restart_button = ttk.Button(self.window, text="Restart", command=self.restart_game)
        self.restart_button.pack(pady=20)

    def play(self, user_choice):
        computer_choice = random.choice(self.choices)
        self.user_choice.set("User choice: " + user_choice)
        self.computer_choice.set("Computer choice: " + computer_choice)

        if user_choice == computer_choice:
            self.result_text.set("It's a tie!")
        elif (user_choice == 'rock' and computer_choice == 'scissors') or \
             (user_choice == 'scissors' and computer_choice == 'paper') or \
             (user_choice == 'paper' and computer_choice == 'rock'):
            self.result_text.set("User wins!")
            self.user_score += 1
        else:
            self.result_text.set("Computer wins!")
            self.computer_score += 1

        self.user_choice.set("User score: " + str(self.user_score))
        self.computer_choice.set("Computer score: " + str(self.computer_score))

    def restart_game(self):
        self.user_score = 0
        self.computer_score = 0
        self.user_choice.set("")
        self.computer_choice.set("")
        self.result_text.set("")

# Usage
root = tk.Tk()
root.title('Rock Paper Scissors Game')
root.geometry('400x200')
game = RockPaperScissorsGame(root)
root.mainloop()
```
